```python
import os
import sys
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import json
```

```python
def load_file(file):
    with open(file) as soup:
        zipcode_file = bs(soup)
    soup.close()
    return zipcode_file

def get_home_data(html_soup):
    removeKeys = (
    'imageLink',
    'shouldHighlight',
    'brokerId',
    'contactPhone',
    'mediumImageLink',
    'hiResImageLink',
    'watchImageLink',
    'contactPhoneExtension',
    'lotId',
    'tvImageLink',
    'tvCollectionImageLink',
    'tvHighResImageLink',
    'zillowHasRightsToImages',
    'desktopWebHdpImageLink',
    'hideZestimate',
    'currency',
    'country'
)
    data_dict = {}
    home_data = html_soup.body.find_all('div', class_ = 'minibubble template hide')
    print(len(home_data))
    for home in range(0, len(home_data)):
        json_load = str(home_data[home].contents[0]).replace("\\", "")
        home_dict = json.loads(json_load)
        #get zipID; returns None if none available
        zpid = home_dict.get('homeInfo', {}).get('zpid')
        if zpid == None:
            continue
        data_dict.update({zpid: {}})
        home_dict.get('homeInfo', {}).pop('zpid', None)
        for key in removeKeys:
            home_dict.get('homeInfo', {}).pop(key, None)
        data_dict[zpid].update(home_dict.get('homeInfo', {}))
    return data_dict

def select_city(city_selection):
    base_path = 'C:/Users/austin.mcdonnell/Projects/Imperial/'
    city_selection = str(city_selection)
    os.chdir(base_path + city_selection + '/ZipCodes/HTML_Data/')
    return os.listdir()

def parse_all_files(city):
    file_list = select_city(city)
    data_dict = {}
    for file in file_list:
        print(file)
        zipcode_data = load_file(file)
        data_dict.update(get_home_data(zipcode_data))
        print('\n')
    return data_dict
```

```python
data = parse_all_files(city='Detroit')
```
